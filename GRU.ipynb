```python
# =========================================================
# uniform 8frame + GRU
# =========================================================
import os, json, random
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np


seed = 42
random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

# Í≤ΩÎ°ú
base_dir = "/content/drive/MyDrive/VIDEO_RETRIEVAL"

video_feats_train_path = os.path.join(base_dir, "clip_video_embeds_train_8f.pt")
video_feats_val_path   = os.path.join(base_dir, "clip_video_embeds_val_8f.pt")
text_feats_train_path  = os.path.join(base_dir, "clip_text_embeds_train_8f.pt")
text_feats_val_path    = os.path.join(base_dir, "clip_text_embeds_val_8f.pt")
save_model_path        = os.path.join(base_dir, "GRU_adapter3_8f_hidden1024.pth")

batch_size = 32
lr = 1e-4
max_epochs = 50
patience = 5
min_delta = 1e-4
dropout = 0.1

# Dataset
class VideoTextFeatureDataset(torch.utils.data.Dataset):
    def __init__(self, video_feats_list, text_feats_list):
        self.video_feats = video_feats_list
        self.text_feats  = text_feats_list
        self.N = len(video_feats_list)
    def __len__(self): return self.N
    def __getitem__(self, idx):
        return self.video_feats[idx], self.text_feats[idx]

# Temporal GRU Adapter
class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        # Attention pooling layer Ï∂îÍ∞Ä
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)             # [B,1,512]
        out, _ = self.gru(x)               # [B,T,H]
        scores = self.attn(out)            # [B,T,1]
        weights = F.softmax(scores, dim=1) # [B,T,1], Í∞Å ÌîÑÎ†àÏûÑÏùò Ï§ëÏöîÎèÑ
        pooled = (out * weights).sum(dim=1) # [B,H], weighted sum
        out = self.proj(pooled)            # projection
        out = self.norm(out)
        return F.normalize(out, dim=-1)



# CLIP-style Contrastive Loss
class CLIPLoss(nn.Module):
    def __init__(self, init_temp=0.07):
        super().__init__()
        self.log_temp = nn.Parameter(torch.log(torch.tensor(init_temp)))

    def forward(self, video_embeds, text_embeds):
        temp = self.log_temp.exp()
        sim = torch.matmul(F.normalize(video_embeds, dim=-1),
                           F.normalize(text_embeds, dim=-1).T) / temp
        labels = torch.arange(sim.size(0), device=sim.device)
        loss = (F.cross_entropy(sim, labels) + F.cross_entropy(sim.T, labels)) / 2
        return loss

# Main Training
def main():
    # Feature load
    print("üì¶ Loading pre-extracted CLIP features...")
    video_embeds_train = torch.load(video_feats_train_path, map_location="cpu")
    text_embeds_train  = torch.load(text_feats_train_path, map_location="cpu")
    video_embeds_val = torch.load(video_feats_val_path, map_location="cpu")
    text_embeds_val  = torch.load(text_feats_val_path, map_location="cpu")

    print(f"Train videos: {len(video_embeds_train)} | Val videos: {len(video_embeds_val)}")

    # Dataset / DataLoader
    train_dataset = VideoTextFeatureDataset(video_embeds_train, text_embeds_train)
    val_dataset   = VideoTextFeatureDataset(video_embeds_val, text_embeds_val)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    # Model
    input_dim = video_embeds_train[0].shape[-1]
    model = TemporalGRU(input_dim=input_dim,
                    hidden_dim=1024,
                    num_layers=1,
                    dropout=dropout).to(device)

    # Learnable Temperature Loss
    clip_loss_fn = CLIPLoss().to(device)

    optimizer = torch.optim.AdamW(
        list(model.parameters()) + list(clip_loss_fn.parameters()),
        lr=lr, weight_decay=1e-4
    )

    # Training Loop
    best_val = float("inf")
    no_improve = 0
    train_losses, val_losses = [], []

    for epoch in range(max_epochs):
        model.train()
        total_train = 0
        for v_batch, t_batch in tqdm(train_loader, desc=f"Train {epoch}", leave=False):
            v, t = v_batch.to(device).float(), t_batch.to(device).float()
            v_out = model(v)
            loss = clip_loss_fn(v_out, t)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_train += loss.item()
        avg_train = total_train / len(train_loader)
        train_losses.append(avg_train)

        # Validation
        model.eval()
        total_val = 0
        with torch.no_grad():
            for v_batch, t_batch in tqdm(val_loader, desc=f"Val {epoch}", leave=False):
                v, t = v_batch.to(device).float(), t_batch.to(device).float()
                v_out = model(v)
                loss = clip_loss_fn(v_out, t)
                total_val += loss.item()
        avg_val = total_val / len(val_loader)
        val_losses.append(avg_val)

        print(f"[Epoch {epoch}] Train {avg_train:.4f} | Val {avg_val:.4f}")

        # Early Stopping
        if avg_val < best_val - min_delta:
            best_val = avg_val
            no_improve = 0
            torch.save(model.state_dict(), save_model_path)
            print(f"‚úÖ Saved model: {save_model_path}")
        else:
            no_improve += 1
            print(f"No improvement ({no_improve}/{patience})")
            if no_improve >= patience:
                print("‚õî Early stopping triggered.")
                break


    # Plot loss
    plt.figure(figsize=(8,5))
    plt.plot(train_losses, label="Train Loss", marker="o")
    plt.plot(val_losses, label="Val Loss", marker="s")
    plt.title("Training vs Validation Loss (GRU Adapter3)")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend(); plt.grid(True)
    plt.show()

    print(f"Best Val Loss: {best_val:.6f}")
    print(f"Model saved to: {save_model_path}")

if __name__ == "__main__":
    main()


#------------Ï∂îÎ°† Î∞è ÏÑ±Îä•Ï≤¥ÌÅ¨------------


device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

base_dir = "/content/drive/MyDrive/VIDEO_RETRIEVAL"

video_feats_test_path = os.path.join(base_dir, "clip_video_embeds_test_8f.pt")
text_feats_test_path  = os.path.join(base_dir, "clip_text_embeds_8f.pt")
model_path            = os.path.join(base_dir, "GRU_adapter3_8f_hidden1024.pth")


class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        # Attention pooling layer Ï∂îÍ∞Ä
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)             # [B,1,512]
        out, _ = self.gru(x)               # [B,T,H]
        scores = self.attn(out)            # [B,T,1]
        weights = F.softmax(scores, dim=1) # [B,T,1], Í∞Å ÌîÑÎ†àÏûÑÏùò Ï§ëÏöîÎèÑ
        pooled = (out * weights).sum(dim=1) # [B,H], weighted sum
        out = self.proj(pooled)            # projection
        out = self.norm(out)
        return F.normalize(out, dim=-1)



def to_2d_tensor(x):
    """xÍ∞Ä [N, D] TensorÎ©¥ Í∑∏ÎåÄÎ°ú, Î¶¨Ïä§Ìä∏Î©¥ stackÌï¥ÏÑú [N, D] TensorÎ°ú."""
    if isinstance(x, torch.Tensor):
        return x
    elif isinstance(x, (list, tuple)):
        x = [t if isinstance(t, torch.Tensor) else torch.tensor(t) for t in x]
        return torch.stack(x, dim=0)
    else:
        raise TypeError(f"Unsupported type for features: {type(x)}")

print("üì¶ Loading pre-extracted CLIP features...")
video_embeds = torch.load(video_feats_test_path, map_location="cpu")
text_embeds  = torch.load(text_feats_test_path, map_location="cpu")

# ÎπÑÎîîÏò§ ÌîºÏ≤ò: Tensor or Î¶¨Ïä§Ìä∏ Ï≤òÎ¶¨
if isinstance(video_embeds, torch.Tensor):
    video_embeds_list = [video_embeds[i] for i in range(video_embeds.size(0))]
else:
    video_embeds_list = video_embeds

# ÌÖçÏä§Ìä∏ ÌîºÏ≤ò Ï†ïÍ∑úÌôî
text_mat = F.normalize(to_2d_tensor(text_embeds).float(), dim=-1)

print(f"‚úÖ Loaded {len(video_embeds_list)} video feats and {text_mat.size(0)} text feats")

# Í∏∏Ïù¥ Î∂àÏùºÏπò Ïãú ÏµúÏÜå Í∏∏Ïù¥Ïóê ÎßûÏ∂∞ ÏûêÎ•¥Í∏∞
N = min(len(video_embeds_list), text_mat.size(0))
if N != len(video_embeds_list) or N != text_mat.size(0):
    print(f"‚ö†Ô∏è Mismatch detected: trimming to {N} pairs")
    video_embeds_list = video_embeds_list[:N]
    text_mat = text_mat[:N]

# Î™®Îç∏ Î°úÎìú
model = TemporalGRU().to(device)
state = torch.load(model_path, map_location=device)
if isinstance(state, dict) and "model_state_dict" in state:
    model.load_state_dict(state["model_state_dict"])
else:
    model.load_state_dict(state)
model.eval()
print(f"‚úÖ Loaded trained GRU model from {model_path}")

# ÎπÑÎîîÏò§ Ïù∏ÏΩîÎî©
batch_size = 64
video_outputs = []

with torch.no_grad():
    for i in tqdm(range(0, len(video_embeds_list), batch_size), desc="GRU Encoding", ncols=100, leave=False):
        batch_seq = video_embeds_list[i:i+batch_size]  # Î¶¨Ïä§Ìä∏ of [T,512]
        v = torch.stack([b if isinstance(b, torch.Tensor) else torch.tensor(b) for b in batch_seq], dim=0)
        v = v.to(device).float()
        v_out = model(v)  # [B, 512]
        video_outputs.append(v_out.cpu())

video_mat = F.normalize(torch.cat(video_outputs, dim=0), dim=-1)
sim = video_mat @ text_mat.T

# Retrieval Metrics
def recall_at_k(ranks, k):
    correct = torch.arange(ranks.size(0)).unsqueeze(1)
    hits = (ranks[:, :k] == correct).any(dim=1).float()
    return hits.mean().item()

def median_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:, 1]
    return pos.float().median().item() + 1

def mean_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:, 1]
    return pos.float().mean().item() + 1

# --- Video‚ÜíText ---
ranks_v2t = sim.argsort(dim=1, descending=True)
R1_v2t, R5_v2t, R10_v2t = recall_at_k(ranks_v2t, 1), recall_at_k(ranks_v2t, 5), recall_at_k(ranks_v2t, 10)
MedR_v2t, MnR_v2t = median_rank(ranks_v2t), mean_rank(ranks_v2t)

# --- Text‚ÜíVideo ---
ranks_t2v = sim.argsort(dim=0, descending=True).T
R1_t2v, R5_t2v, R10_t2v = recall_at_k(ranks_t2v, 1), recall_at_k(ranks_t2v, 5), recall_at_k(ranks_t2v, 10)
MedR_t2v, MnR_t2v = median_rank(ranks_t2v), mean_rank(ranks_t2v)


print("\nüéØ Retrieval Performance (GRU Adapter3, 8f)")
print("‚îÄ‚îÄ Video ‚Üí Text ‚îÄ‚îÄ")
print(f"R@1 : {R1_v2t*100:.2f}%")
print(f"R@5 : {R5_v2t*100:.2f}%")
print(f"R@10: {R10_v2t*100:.2f}%")
print(f"MedR: {MedR_v2t:.1f}")
print(f"MnR : {MnR_v2t:.1f}")

print("\n‚îÄ‚îÄ Text ‚Üí Video ‚îÄ‚îÄ")
print(f"R@1 : {R1_t2v*100:.2f}%")
print(f"R@5 : {R5_t2v*100:.2f}%")
print(f"R@10: {R10_t2v*100:.2f}%")
print(f"MedR: {MedR_t2v:.1f}")
print(f"MnR : {MnR_t2v:.1f}")
```

```python
# =========================================================
# Random 8frame + GRU
# =========================================================
import os, json, random
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np

seed = 42
random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

# Í≤ΩÎ°ú (Ïù¥ÎØ∏ Ï°¥Ïû¨ÌïòÎäî ÌîºÏ≤ò)
base_dir = "/content/drive/MyDrive/VIDEO_RETRIEVAL"

video_feats_train_path = os.path.join(base_dir, "clip_video_embeds_train_random8.pt")
video_feats_val_path   = os.path.join(base_dir, "clip_video_embeds_val_random8.pt")
text_feats_train_path  = os.path.join(base_dir, "clip_text_embeds_train_random8.pt")
text_feats_val_path    = os.path.join(base_dir, "clip_text_embeds_val_random8.pt")
save_model_path        = os.path.join(base_dir, "GRU_adapter3_random8f_hidden1024.pth")


batch_size = 32
lr = 1e-4
max_epochs = 50
patience = 5
min_delta = 1e-4
dropout = 0.1

# Dataset
class VideoTextFeatureDataset(torch.utils.data.Dataset):
    def __init__(self, video_feats_list, text_feats_list):
        self.video_feats = video_feats_list
        self.text_feats  = text_feats_list
        self.N = len(video_feats_list)
    def __len__(self): return self.N
    def __getitem__(self, idx):
        return self.video_feats[idx], self.text_feats[idx]

# Temporal GRU Adapter
class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        # Attention pooling layer Ï∂îÍ∞Ä
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)             # [B,1,512]
        out, _ = self.gru(x)               # [B,T,H]
        scores = self.attn(out)            # [B,T,1]
        weights = F.softmax(scores, dim=1) # [B,T,1], Í∞Å ÌîÑÎ†àÏûÑÏùò Ï§ëÏöîÎèÑ
        pooled = (out * weights).sum(dim=1) # [B,H], weighted sum
        out = self.proj(pooled)            # projection
        out = self.norm(out)
        return F.normalize(out, dim=-1)



# CLIP-style Contrastive Loss
class CLIPLoss(nn.Module):
    def __init__(self, init_temp=0.07):
        super().__init__()
        self.log_temp = nn.Parameter(torch.log(torch.tensor(init_temp)))

    def forward(self, video_embeds, text_embeds):
        temp = self.log_temp.exp()
        sim = torch.matmul(F.normalize(video_embeds, dim=-1),
                           F.normalize(text_embeds, dim=-1).T) / temp
        labels = torch.arange(sim.size(0), device=sim.device)
        loss = (F.cross_entropy(sim, labels) + F.cross_entropy(sim.T, labels)) / 2
        return loss

# Main Training
def main():
    # Feature load
    print("üì¶ Loading pre-extracted CLIP features...")
    video_embeds_train = torch.load(video_feats_train_path, map_location="cpu")
    text_embeds_train  = torch.load(text_feats_train_path, map_location="cpu")
    video_embeds_val = torch.load(video_feats_val_path, map_location="cpu")
    text_embeds_val  = torch.load(text_feats_val_path, map_location="cpu")

    print(f"Train videos: {len(video_embeds_train)} | Val videos: {len(video_embeds_val)}")

    # Dataset / DataLoader
    train_dataset = VideoTextFeatureDataset(video_embeds_train, text_embeds_train)
    val_dataset   = VideoTextFeatureDataset(video_embeds_val, text_embeds_val)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    # Model
    input_dim = video_embeds_train[0].shape[-1]
    model = TemporalGRU(input_dim=input_dim,
                    hidden_dim=1024,
                    num_layers=1,
                    dropout=dropout).to(device)

    # Learnable Temperature Loss
    clip_loss_fn = CLIPLoss().to(device)

    optimizer = torch.optim.AdamW(
        list(model.parameters()) + list(clip_loss_fn.parameters()),
        lr=lr, weight_decay=1e-4
    )

    # Training Loop
    best_val = float("inf")
    no_improve = 0
    train_losses, val_losses = [], []

    for epoch in range(max_epochs):
        model.train()
        total_train = 0
        for v_batch, t_batch in tqdm(train_loader, desc=f"Train {epoch}", leave=False):
            v, t = v_batch.to(device).float(), t_batch.to(device).float()
            v_out = model(v)
            loss = clip_loss_fn(v_out, t)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_train += loss.item()
        avg_train = total_train / len(train_loader)
        train_losses.append(avg_train)

        # Validation
        model.eval()
        total_val = 0
        with torch.no_grad():
            for v_batch, t_batch in tqdm(val_loader, desc=f"Val {epoch}", leave=False):
                v, t = v_batch.to(device).float(), t_batch.to(device).float()
                v_out = model(v)
                loss = clip_loss_fn(v_out, t)
                total_val += loss.item()
        avg_val = total_val / len(val_loader)
        val_losses.append(avg_val)

        print(f"[Epoch {epoch}] Train {avg_train:.4f} | Val {avg_val:.4f}")

        # Early Stopping
        if avg_val < best_val - min_delta:
            best_val = avg_val
            no_improve = 0
            torch.save(model.state_dict(), save_model_path)
            print(f"‚úÖ Saved model: {save_model_path}")
        else:
            no_improve += 1
            print(f"No improvement ({no_improve}/{patience})")
            if no_improve >= patience:
                print("‚õî Early stopping triggered.")
                break


    # Plot loss
    plt.figure(figsize=(8,5))
    plt.plot(train_losses, label="Train Loss", marker="o")
    plt.plot(val_losses, label="Val Loss", marker="s")
    plt.title("Training vs Validation Loss (GRU Adapter3)")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend(); plt.grid(True)
    plt.show()

    print(f"Best Val Loss: {best_val:.6f}")
    print(f"Model saved to: {save_model_path}")

if __name__ == "__main__":
    main()


#------------------- Ï∂îÎ°† Î∞è ÏÑ±Îä• ÌôïÏù∏ ------------------------------


device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

base_dir = "/content/drive/MyDrive/VIDEO_RETRIEVAL"

video_feats_test_path = os.path.join(base_dir, "clip_video_embeds_test_random8.pt")
text_feats_test_path  = os.path.join(base_dir, "clip_text_embeds_test_random8.pt")
model_path            = os.path.join(base_dir, "GRU_adapter3_random8f_hidden1024.pth")


class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        # Attention pooling layer Ï∂îÍ∞Ä
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)             # [B,1,512]
        out, _ = self.gru(x)               # [B,T,H]
        scores = self.attn(out)            # [B,T,1]
        weights = F.softmax(scores, dim=1) # [B,T,1], Í∞Å ÌîÑÎ†àÏûÑÏùò Ï§ëÏöîÎèÑ
        pooled = (out * weights).sum(dim=1) # [B,H], weighted sum
        out = self.proj(pooled)            # projection
        out = self.norm(out)
        return F.normalize(out, dim=-1)


def to_2d_tensor(x):
    """xÍ∞Ä [N, D] TensorÎ©¥ Í∑∏ÎåÄÎ°ú, Î¶¨Ïä§Ìä∏Î©¥ stackÌï¥ÏÑú [N, D] TensorÎ°ú."""
    if isinstance(x, torch.Tensor):
        return x
    elif isinstance(x, (list, tuple)):
        x = [t if isinstance(t, torch.Tensor) else torch.tensor(t) for t in x]
        return torch.stack(x, dim=0)
    else:
        raise TypeError(f"Unsupported type for features: {type(x)}")

# ÌîºÏ≤ò Î°úÎìú
print("üì¶ Loading pre-extracted CLIP features...")
video_embeds = torch.load(video_feats_test_path, map_location="cpu")
text_embeds  = torch.load(text_feats_test_path, map_location="cpu")

# ÎπÑÎîîÏò§ ÌîºÏ≤ò: Tensor or Î¶¨Ïä§Ìä∏ Ï≤òÎ¶¨
if isinstance(video_embeds, torch.Tensor):
    video_embeds_list = [video_embeds[i] for i in range(video_embeds.size(0))]
else:
    video_embeds_list = video_embeds

# ÌÖçÏä§Ìä∏ ÌîºÏ≤ò Ï†ïÍ∑úÌôî
text_mat = F.normalize(to_2d_tensor(text_embeds).float(), dim=-1)

print(f"‚úÖ Loaded {len(video_embeds_list)} video feats and {text_mat.size(0)} text feats")

# Í∏∏Ïù¥ Î∂àÏùºÏπò Ïãú ÏµúÏÜå Í∏∏Ïù¥Ïóê ÎßûÏ∂∞ ÏûêÎ•¥Í∏∞
N = min(len(video_embeds_list), text_mat.size(0))
if N != len(video_embeds_list) or N != text_mat.size(0):
    print(f" Mismatch detected: trimming to {N} pairs")
    video_embeds_list = video_embeds_list[:N]
    text_mat = text_mat[:N]

# Î™®Îç∏ Î°úÎìú
model = TemporalGRU().to(device)
state = torch.load(model_path, map_location=device)
if isinstance(state, dict) and "model_state_dict" in state:
    model.load_state_dict(state["model_state_dict"])
else:
    model.load_state_dict(state)
model.eval()
print(f"‚úÖ Loaded trained GRU model from {model_path}")

# ÎπÑÎîîÏò§ Ïù∏ÏΩîÎî©
batch_size = 64
video_outputs = []

with torch.no_grad():
    for i in tqdm(range(0, len(video_embeds_list), batch_size), desc="GRU Encoding", ncols=100, leave=False):
        batch_seq = video_embeds_list[i:i+batch_size]  # Î¶¨Ïä§Ìä∏ of [T,512]
        v = torch.stack([b if isinstance(b, torch.Tensor) else torch.tensor(b) for b in batch_seq], dim=0)
        v = v.to(device).float()
        v_out = model(v)  # [B, 512]
        video_outputs.append(v_out.cpu())

video_mat = F.normalize(torch.cat(video_outputs, dim=0), dim=-1)
sim = video_mat @ text_mat.T

# Retrieval Metrics
def recall_at_k(ranks, k):
    correct = torch.arange(ranks.size(0)).unsqueeze(1)
    hits = (ranks[:, :k] == correct).any(dim=1).float()
    return hits.mean().item()

def median_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:, 1]
    return pos.float().median().item() + 1

def mean_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:, 1]
    return pos.float().mean().item() + 1

# --- Video‚ÜíText ---
ranks_v2t = sim.argsort(dim=1, descending=True)
R1_v2t, R5_v2t, R10_v2t = recall_at_k(ranks_v2t, 1), recall_at_k(ranks_v2t, 5), recall_at_k(ranks_v2t, 10)
MedR_v2t, MnR_v2t = median_rank(ranks_v2t), mean_rank(ranks_v2t)

# --- Text‚ÜíVideo ---
ranks_t2v = sim.argsort(dim=0, descending=True).T
R1_t2v, R5_t2v, R10_t2v = recall_at_k(ranks_t2v, 1), recall_at_k(ranks_t2v, 5), recall_at_k(ranks_t2v, 10)
MedR_t2v, MnR_t2v = median_rank(ranks_t2v), mean_rank(ranks_t2v)

print("\nüéØ Retrieval Performance (GRU Adapter3, 32f)")
print("‚îÄ‚îÄ Video ‚Üí Text ‚îÄ‚îÄ")
print(f"R@1 : {R1_v2t*100:.2f}%")
print(f"R@5 : {R5_v2t*100:.2f}%")
print(f"R@10: {R10_v2t*100:.2f}%")
print(f"MedR: {MedR_v2t:.1f}")
print(f"MnR : {MnR_v2t:.1f}")

print("\n‚îÄ‚îÄ Text ‚Üí Video ‚îÄ‚îÄ")
print(f"R@1 : {R1_t2v*100:.2f}%")
print(f"R@5 : {R5_t2v*100:.2f}%")
print(f"R@10: {R10_t2v*100:.2f}%")
print(f"MedR: {MedR_t2v:.1f}")
print(f"MnR : {MnR_t2v:.1f}")
```

```python
# =========================================================
# Random 16frame + GRU
# =========================================================

import os, json, random, re, warnings
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
from concurrent.futures import ThreadPoolExecutor

seed = 42
random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

train_frames_root = "/content/drive/MyDrive/AY/random/Trainframes_random16"
val_frames_root   = "/content/drive/MyDrive/AY/random/Valframes_random16"

captions_train_path = "/content/drive/MyDrive/AY/random/captions_train.json"
captions_val_path   = "/content/drive/MyDrive/AY/random/captions_val.json"

video_feats_train_path = "/content/drive/MyDrive/AY/random/clip_video_embeds_train_random16_fixed.pt"
video_feats_val_path   = "/content/drive/MyDrive/AY/random/clip_video_embeds_val_random16_fixed.pt"
text_feats_train_path  = "/content/drive/MyDrive/AY/random/clip_text_embeds_train_random16_fixed.pt"
text_feats_val_path    = "/content/drive/MyDrive/AY/random/clip_text_embeds_val_random16_fixed.pt"
save_model_path        = "/content/drive/MyDrive/AY/random/gru_adapter_random16_fixed.pth"

# ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞
batch_size = 32
lr = 1e-4
max_epochs = 50
patience = 5
min_delta = 1e-4
dropout = 0.1
workers = 8  # for parallel image loading

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# Î≥ëÎ†¨ Ïù¥ÎØ∏ÏßÄ Î°úÎìú Ìï®Ïàò
def load_images_parallel(frame_paths, max_workers=8):
    def _load_one(path):
        try:
            return Image.open(path).convert("RGB")
        except:
            return None
    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        imgs = list(ex.map(_load_one, frame_paths))
    return [i for i in imgs if i is not None]

# CLIP ÌîºÏ≤ò Ï∂îÏ∂ú
@torch.no_grad()
def extract_video_and_text_features_fast(frames_root, captions_dict,
                                         video_out_path, text_out_path,
                                         clip_model, clip_processor,
                                         max_workers=8,
                                         batch_size=8):
    if os.path.exists(video_out_path) and os.path.exists(text_out_path):
        print(f"[SKIP] Features already exist: {video_out_path}, {text_out_path}")
        return

    clip_model.eval()
    video_embeds, text_embeds = [], []
    video_names = sorted([d for d in os.listdir(frames_root)
                          if os.path.isdir(os.path.join(frames_root, d))])
    print(f"‚ö° Fast feature extraction: {len(video_names)} videos from {frames_root}")

    for vid in tqdm(video_names, total=len(video_names),
                    desc="Extracting CLIP features", leave=False, ncols=100):

        vid_path = os.path.join(frames_root, vid)
        frame_paths = sorted([
            os.path.join(vid_path, f)
            for f in os.listdir(vid_path)
            if f.lower().endswith(('.jpg', '.jpeg', '.png'))
        ])
        if not frame_paths:
            continue

        # ÌîÑÎ†àÏûÑ Îã®ÏúÑ ÌîºÏ≤ò Ï∂îÏ∂ú
        frame_feats = []
        for i in range(0, len(frame_paths), batch_size):
            batch_paths = frame_paths[i:i+batch_size]
            imgs = load_images_parallel(batch_paths, max_workers=max_workers)
            if not imgs:
                continue

            inputs = clip_processor(images=imgs, return_tensors="pt").to(device, non_blocking=True)
            feats = clip_model.get_image_features(**inputs)
            feats = F.normalize(feats, dim=-1).detach().cpu()
            frame_feats.append(feats)

        if not frame_feats:
            continue

        feats = torch.cat(frame_feats, dim=0)
        video_embeds.append(feats)

        # video1234 ‚Üí "1234"Î°ú key Ï†ïÍ∑úÌôî
        caption_key = re.sub(r"[^0-9]", "", vid)
        caption_text = captions_dict.get(caption_key, "")

        if isinstance(caption_text, list):
            caption_text = caption_text[0]
        elif not isinstance(caption_text, str):
            caption_text = str(caption_text)

        text_inputs = clip_processor(
            text=caption_text, return_tensors="pt", truncation=True
        ).to(device, non_blocking=True)
        tfeat = clip_model.get_text_features(**text_inputs)
        tfeat = F.normalize(tfeat, dim=-1).squeeze(0).detach().cpu()
        text_embeds.append(tfeat)

    torch.save(video_embeds, video_out_path)
    torch.save(text_embeds, text_out_path)
    print(f"‚úÖ Saved features ‚Üí {video_out_path}, {text_out_path}")

# Dataset Ï†ïÏùò
class VideoTextFeatureDataset(torch.utils.data.Dataset):
    def __init__(self, video_feats_list, text_feats_list):
        self.video_feats = video_feats_list
        self.text_feats  = text_feats_list
        self.N = len(video_feats_list)
    def __len__(self): return self.N
    def __getitem__(self, idx):
        return self.video_feats[idx], self.text_feats[idx]

# Temporal GRU Adapter
class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)
        out, _ = self.gru(x)
        scores = self.attn(out)
        weights = F.softmax(scores, dim=1)
        pooled = (out * weights).sum(dim=1)
        out = self.proj(pooled)
        out = self.norm(out)
        return F.normalize(out, dim=-1)

# CLIP-style Contrastive Loss (Learnable Temperature)
class CLIPLoss(nn.Module):
    def __init__(self, init_temp=0.07):
        super().__init__()
        self.log_temp = nn.Parameter(torch.log(torch.tensor(init_temp)))

    def forward(self, video_embeds, text_embeds):
        temp = self.log_temp.exp()
        sim = torch.matmul(F.normalize(video_embeds, dim=-1),
                           F.normalize(text_embeds, dim=-1).T) / temp
        labels = torch.arange(sim.size(0), device=sim.device)
        loss = (F.cross_entropy(sim, labels) + F.cross_entropy(sim.T, labels)) / 2
        return loss

# Main Training Loop
def main():
    clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16").to(device)
    clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")

    with open(captions_train_path, "r") as f:
        captions_train = json.load(f)
    with open(captions_val_path, "r") as f:
        captions_val = json.load(f)

    extract_video_and_text_features_fast(train_frames_root, captions_train,
                                         video_feats_train_path, text_feats_train_path,
                                         clip_model, clip_processor, max_workers=workers)
    extract_video_and_text_features_fast(val_frames_root, captions_val,
                                         video_feats_val_path, text_feats_val_path,
                                         clip_model, clip_processor, max_workers=workers)

    print("üì¶ Loading pre-extracted CLIP features...")
    video_embeds_train = torch.load(video_feats_train_path, map_location="cpu")
    text_embeds_train  = torch.load(text_feats_train_path, map_location="cpu")
    video_embeds_val   = torch.load(video_feats_val_path, map_location="cpu")
    text_embeds_val    = torch.load(text_feats_val_path, map_location="cpu")

    print(f"Train videos: {len(video_embeds_train)} | Val videos: {len(video_embeds_val)}")

    train_dataset = VideoTextFeatureDataset(video_embeds_train, text_embeds_train)
    val_dataset   = VideoTextFeatureDataset(video_embeds_val, text_embeds_val)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    input_dim = video_embeds_train[0].shape[-1]
    model = TemporalGRU(input_dim=input_dim, hidden_dim=1024,
                        num_layers=1, dropout=dropout).to(device)
    clip_loss_fn = CLIPLoss().to(device)

    optimizer = torch.optim.AdamW(
        list(model.parameters()) + list(clip_loss_fn.parameters()),
        lr=lr, weight_decay=1e-4
    )

    best_val = float("inf")
    no_improve = 0
    train_losses, val_losses = [], []

    for epoch in range(max_epochs):
        model.train()
        total_train = 0
        for v_batch, t_batch in tqdm(train_loader, desc=f"Train {epoch}", leave=False):
            v, t = v_batch.to(device).float(), t_batch.to(device).float()
            v_out = model(v)
            loss = clip_loss_fn(v_out, t)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_train += loss.item()
        avg_train = total_train / len(train_loader)
        train_losses.append(avg_train)

        model.eval()
        total_val = 0
        with torch.no_grad():
            for v_batch, t_batch in tqdm(val_loader, desc=f"Val {epoch}", leave=False):
                v, t = v_batch.to(device).float(), t_batch.to(device).float()
                v_out = model(v)
                loss = clip_loss_fn(v_out, t)
                total_val += loss.item()
        avg_val = total_val / len(val_loader)
        val_losses.append(avg_val)

        print(f"[Epoch {epoch}] Train {avg_train:.4f} | Val {avg_val:.4f}")

        if avg_val < best_val - min_delta:
            best_val = avg_val
            no_improve = 0
            torch.save(model.state_dict(), save_model_path)
            print(f"‚úÖ Saved model: {save_model_path}")
        else:
            no_improve += 1
            print(f"No improvement ({no_improve}/{patience})")
            if no_improve >= patience:
                print("‚õî Early stopping triggered.")
                break

    plt.figure(figsize=(8,5))
    plt.plot(train_losses, label="Train Loss", marker="o")
    plt.plot(val_losses, label="Val Loss", marker="s")
    plt.title("Training vs Validation Loss (GRU Adapter Random16, Fixed)")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend(); plt.grid(True)
    plt.show()

    print(f"Best Val Loss: {best_val:.6f}")
    print(f"Model saved to: {save_model_path}")

if __name__ == "__main__":
    main()
```

```python
# -------------- Random 16frame + GRU (Ï∂îÎ°† Î∞è ÏÑ±Îä• ÌôïÏù∏) ------------------------------


import os, json, re, torch, torch.nn as nn, torch.nn.functional as F
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
import warnings

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)


base_dir = "/content/drive/MyDrive/AY/random"
test_frames_root   = os.path.join(base_dir, "Test_1kA_random16")
captions_test_path = os.path.join(base_dir, "captions_test_1k.json")

video_feats_test_path = os.path.join(base_dir, "clip_video_embeds_test_random16_fixed.pt")
text_feats_test_path  = os.path.join(base_dir, "clip_text_embeds_test_random16_fixed.pt")
model_path            = os.path.join(base_dir, "gru_adapter_random16_fixed.pth")

def load_images_parallel(frame_paths, max_workers=8):
    def _load_one(path):
        try:
            return Image.open(path).convert("RGB")
        except:
            return None
    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        imgs = list(ex.map(_load_one, frame_paths))
    return [i for i in imgs if i is not None]

# CLIP Î™®Îç∏ Î°úÎìú
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")
clip_model.eval()

# Í∏∞Ï°¥ ÌîºÏ≤ò Î°úÎìú
if os.path.exists(video_feats_test_path):
    existing_video_embeds = torch.load(video_feats_test_path, map_location="cpu")
    print(f" Loaded existing video features: {len(existing_video_embeds)}")
else:
    existing_video_embeds = []

if os.path.exists(text_feats_test_path):
    existing_text_embeds = torch.load(text_feats_test_path, map_location="cpu")
    print(f" Loaded existing text features: {len(existing_text_embeds)}")
else:
    existing_text_embeds = []

# Í∏∞Ï°¥ ÎπÑÎîîÏò§ Ïù¥Î¶ÑÎì§ Î°úÎìú
existing_folders = set(
    sorted([d for d in os.listdir(test_frames_root)
            if os.path.isdir(os.path.join(test_frames_root, d))])
)
print(f" Total video folders: {len(existing_folders)}")

# Ïù¥ÎØ∏ Ï≤òÎ¶¨Îêú ÎπÑÎîîÏò§ Ïàò Í∞êÏßÄ (ÏûÑÎ≤†Îî© Í∞úÏàòÏôÄ Ìè¥Îçî Í∞úÏàò ÎπÑÍµê)
processed_count = len(existing_video_embeds)
if processed_count < len(existing_folders):
    print(f" Missing videos detected: {len(existing_folders) - processed_count}")
else:
    print(" All videos already processed.")

# ÎàÑÎùΩÎêú ÎπÑÎîîÏò§Îßå Ï∂îÏ∂ú
@torch.no_grad()
def extract_missing_features(frames_root, captions_path,
                             clip_model, clip_processor,
                             existing_count,
                             video_out_path, text_out_path):
    with open(captions_path, "r") as f:
        captions_dict = json.load(f)

    video_embeds, text_embeds = [], []
    video_folders = sorted([d for d in os.listdir(frames_root)
                            if os.path.isdir(os.path.join(frames_root, d))])

    missing_videos = video_folders[existing_count:]
    print(f" Extracting features for {len(missing_videos)} missing videos...")

    for vid in tqdm(missing_videos, desc="Extracting missing video feats", ncols=100):
        vid_path = os.path.join(frames_root, vid)
        frame_files = sorted([os.path.join(vid_path, f)
                              for f in os.listdir(vid_path)
                              if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
        if len(frame_files) == 0:
            continue

        imgs = load_images_parallel(frame_files)
        if len(imgs) == 0:
            continue

        inputs = clip_processor(images=imgs, return_tensors="pt").to(device)
        feats = clip_model.get_image_features(**inputs)
        feats = F.normalize(feats, dim=-1).cpu()
        video_embeds.append(feats)

        key = re.sub(r"[^0-9]", "", vid)
        cap = captions_dict.get(key, None)
        if cap is None:
            cap = ""
        if isinstance(cap, list):
            cap = cap[0]
        elif not isinstance(cap, str):
            cap = str(cap)

        t_inputs = clip_processor(text=cap, return_tensors="pt", truncation=True).to(device)
        t_feats = clip_model.get_text_features(**t_inputs)
        t_feats = F.normalize(t_feats, dim=-1).squeeze(0).cpu()
        text_embeds.append(t_feats)

    # Í∏∞Ï°¥ + ÏÉàÎ°ú Ï∂îÍ∞ÄÎêú ÌîºÏ≤ò Ìï©ÏπòÍ∏∞
    if os.path.exists(video_out_path):
        old_videos = torch.load(video_out_path, map_location="cpu")
        video_embeds = old_videos + video_embeds
    if os.path.exists(text_out_path):
        old_texts = torch.load(text_out_path, map_location="cpu")
        text_embeds = old_texts + text_embeds

    torch.save(video_embeds, video_out_path)
    torch.save(text_embeds, text_out_path)
    print(f" Updated features saved: {len(video_embeds)} videos total")

extract_missing_features(
    test_frames_root,
    captions_test_path,
    clip_model,
    clip_processor,
    len(existing_video_embeds),
    video_feats_test_path,
    text_feats_test_path
)

# GRU Î™®Îç∏ Ï†ïÏùò
class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)
        out, _ = self.gru(x)
        scores = self.attn(out)
        weights = F.softmax(scores, dim=1)
        pooled = (out * weights).sum(dim=1)
        out = self.proj(pooled)
        out = self.norm(out)
        return F.normalize(out, dim=-1)

# GRU Î™®Îç∏ Î°úÎìú
model = TemporalGRU().to(device)
state = torch.load(model_path, map_location=device)
model.load_state_dict(state)
model.eval()
print(f" Loaded trained GRU model from {model_path}")

# ÎπÑÎîîÏò§ Ïù∏ÏΩîÎî©
batch_size = 64
video_outputs = []

with torch.no_grad():
    for i in tqdm(range(0, len(video_embeds), batch_size), desc="GRU Encoding", ncols=100, leave=False):
        batch = video_embeds[i:i+batch_size]
        v = torch.stack(batch).to(device).float()
        v_out = model(v)
        video_outputs.append(v_out.cpu())

video_mat = torch.cat(video_outputs, dim=0)
text_mat  = F.normalize(torch.stack(text_embeds), dim=-1)

# Retrieval ÌèâÍ∞Ä (Video‚ÜíText, Text‚ÜíVideo)
video_mat = F.normalize(video_mat, dim=-1)
text_mat  = F.normalize(text_mat, dim=-1)
sim = video_mat @ text_mat.T

def recall_at_k(ranks, k):
    correct = torch.arange(ranks.size(0)).unsqueeze(1)
    hits = (ranks[:, :k] == correct).any(dim=1).float()
    return hits.mean().item()

def median_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:,1]
    return pos.float().median().item() + 1

def mean_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:,1]
    return pos.float().mean().item() + 1

# --- Video‚ÜíText ---
ranks_v2t = sim.argsort(dim=1, descending=True)
R1_v2t, R5_v2t, R10_v2t = recall_at_k(ranks_v2t,1), recall_at_k(ranks_v2t,5), recall_at_k(ranks_v2t,10)
MedR_v2t, MnR_v2t = median_rank(ranks_v2t), mean_rank(ranks_v2t)

# --- Text‚ÜíVideo ---
ranks_t2v = sim.argsort(dim=0, descending=True).T
R1_t2v, R5_t2v, R10_t2v = recall_at_k(ranks_t2v,1), recall_at_k(ranks_t2v,5), recall_at_k(ranks_t2v,10)
MedR_t2v, MnR_t2v = median_rank(ranks_t2v), mean_rank(ranks_t2v)

print("\n Retrieval Performance (Random 16f, GRU-based)")
print("‚îÄ‚îÄ Video ‚Üí Text ‚îÄ‚îÄ")
print(f"R@1 : {R1_v2t*100:.2f}%")
print(f"R@5 : {R5_v2t*100:.2f}%")
print(f"R@10: {R10_v2t*100:.2f}%")
print(f"MedR: {MedR_v2t:.1f}")
print(f"MnR : {MnR_v2t:.1f}")

print("\n‚îÄ‚îÄ Text ‚Üí Video ‚îÄ‚îÄ")
print(f"R@1 : {R1_t2v*100:.2f}%")
print(f"R@5 : {R5_t2v*100:.2f}%")
print(f"R@10: {R10_t2v*100:.2f}%")
print(f"MedR: {MedR_t2v:.1f}")
print(f"MnR : {MnR_t2v:.1f}")
```

```python
# =========================================================
# Random 32frame + GRU
# =========================================================

import os, json, random, re, warnings
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
from concurrent.futures import ThreadPoolExecutor

seed = 42
random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

train_frames_root = "/content/drive/MyDrive/VIDEO_RETRIEVAL/Trainframes_random32"
val_frames_root   = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/Valframes_random32"

captions_train_path = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/captions_train.json"
captions_val_path   = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/captions_val.json"

video_feats_train_path = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/clip_video_embeds_train_random32_fixed.pt"
video_feats_val_path   = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/clip_video_embeds_val_random32_fixed.pt"
text_feats_train_path  = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/clip_text_embeds_train_random32_fixed.pt"
text_feats_val_path    = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/clip_text_embeds_val_random32_fixed.pt"
save_model_path        = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random/gru_adapter_random32_fixed.pth"

# ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞
batch_size = 32
lr = 1e-4
max_epochs = 50
patience = 5
min_delta = 1e-4
dropout = 0.1
workers = 8

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# Î≥ëÎ†¨ Ïù¥ÎØ∏ÏßÄ Î°úÎìú Ìï®Ïàò
def load_images_parallel(frame_paths, max_workers=8):
    def _load_one(path):
        try:
            return Image.open(path).convert("RGB")
        except:
            return None
    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        imgs = list(ex.map(_load_one, frame_paths))
    return [i for i in imgs if i is not None]

# ÎàÑÎùΩÎêú / ÏßßÏùÄ ÎπÑÎîîÏò§Îßå Îã§Ïãú Ï∂îÏ∂ú
@torch.no_grad()
def reextract_missing_or_short(frames_root, captions_dict,
                               video_out_path, text_out_path,
                               clip_model, clip_processor,
                               min_frames=32,
                               max_workers=8, batch_size=8):
    """Ïù¥ÎØ∏ Ï†ÄÏû•Îêú pt ÌååÏùº Î∂àÎü¨ÏôÄÏÑú ÎàÑÎùΩ or ÏßßÏùÄ Ìï≠Î™©Îßå Ïû¨Ï∂îÏ∂ú."""
    clip_model.eval()
    video_embeds, text_embeds = [], []
    existing_videos = set()

    # Í∏∞Ï°¥ feature Î∂àÎü¨Ïò§Í∏∞
    if os.path.exists(video_out_path) and os.path.exists(text_out_path):
        video_embeds = torch.load(video_out_path, map_location="cpu")
        text_embeds  = torch.load(text_out_path, map_location="cpu")
        existing_videos = {f"video{i}" for i in range(len(video_embeds))}
        print(f"üì¶ Loaded existing features ({len(existing_videos)} videos)")
    else:
        print("üöÄ No existing features found ‚Äî full extraction will be done.")

    video_names = sorted([d for d in os.listdir(frames_root)
                          if os.path.isdir(os.path.join(frames_root, d))])
    print(f"‚ö° Checking {len(video_names)} videos from {frames_root}")

    new_video_embeds, new_text_embeds = [], []
    reprocessed = 0

    for vid in tqdm(video_names, total=len(video_names), desc="Re-extracting", ncols=100):
        vid_path = os.path.join(frames_root, vid)
        frame_paths = sorted([
            os.path.join(vid_path, f)
            for f in os.listdir(vid_path)
            if f.lower().endswith(('.jpg', '.jpeg', '.png'))
        ])

        # ÏôÑÏ†Ñ Îπà Ìè¥Îçî
        if len(frame_paths) == 0:
            continue

        idx = int(re.sub(r"[^0-9]", "", vid)) if re.search(r"\d+", vid) else None
        need_reprocess = True

        if idx is not None and idx < len(video_embeds):
            existing_feat = video_embeds[idx]
            if isinstance(existing_feat, torch.Tensor) and existing_feat.size(0) >= min_frames:
                need_reprocess = False

        if not need_reprocess:
            continue

        # Îã§Ïãú Ï∂îÏ∂ú
        frame_feats = []
        for i in range(0, len(frame_paths), batch_size):
            batch_paths = frame_paths[i:i+batch_size]
            imgs = load_images_parallel(batch_paths, max_workers=max_workers)
            if not imgs:
                continue
            inputs = clip_processor(images=imgs, return_tensors="pt").to(device)
            feats = clip_model.get_image_features(**inputs)
            feats = F.normalize(feats, dim=-1).detach().cpu()
            frame_feats.append(feats)
        if not frame_feats:
            continue
        feats = torch.cat(frame_feats, dim=0)

        caption_key = re.sub(r"[^0-9]", "", vid)
        caption_text = captions_dict.get(caption_key, "")
        if isinstance(caption_text, list):
            caption_text = caption_text[0]
        elif not isinstance(caption_text, str):
            caption_text = str(caption_text)

        text_inputs = clip_processor(text=caption_text, return_tensors="pt", truncation=True).to(device)
        tfeat = clip_model.get_text_features(**text_inputs)
        tfeat = F.normalize(tfeat, dim=-1).squeeze(0).detach().cpu()

        # Í∏∞Ï°¥ featureÏóê ÎçÆÏñ¥Ïì∞Í∏∞
        if idx is not None and idx < len(video_embeds):
            video_embeds[idx] = feats
            text_embeds[idx] = tfeat
        else:
            video_embeds.append(feats)
            text_embeds.append(tfeat)

        reprocessed += 1

    torch.save(video_embeds, video_out_path)
    torch.save(text_embeds, text_out_path)
    print(f"‚úÖ Re-saved features ({reprocessed} reprocessed videos) ‚Üí {video_out_path}")

# Dataset Ï†ïÏùò
class VideoTextFeatureDataset(torch.utils.data.Dataset):
    def __init__(self, video_feats_list, text_feats_list):
        self.video_feats = video_feats_list
        self.text_feats  = text_feats_list
    def __len__(self): return len(self.video_feats)
    def __getitem__(self, idx):
        return self.video_feats[idx], self.text_feats[idx]

def collate_fn(batch):
    videos, texts = zip(*batch)
    videos_padded = pad_sequence(videos, batch_first=True)
    texts_stacked = torch.stack(texts)
    return videos_padded, texts_stacked

# Temporal GRU Adapter
class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(input_dim, hidden_dim, num_layers=num_layers,
                          batch_first=True, dropout=dropout if num_layers > 1 else 0.0)
        self.attn = nn.Sequential(nn.Linear(hidden_dim, 128), nn.Tanh(), nn.Linear(128, 1))
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)
    def forward(self, x):
        out, _ = self.gru(x)
        w = F.softmax(self.attn(out), dim=1)
        pooled = (out * w).sum(dim=1)
        return F.normalize(self.norm(self.proj(pooled)), dim=-1)

# CLIP Contrastive Loss
class CLIPLoss(nn.Module):
    def __init__(self, init_temp=0.07):
        super().__init__()
        self.log_temp = nn.Parameter(torch.log(torch.tensor(init_temp)))
    def forward(self, v, t):
        temp = self.log_temp.exp()
        sim = (F.normalize(v, dim=-1) @ F.normalize(t, dim=-1).T) / temp
        lbl = torch.arange(sim.size(0), device=sim.device)
        return (F.cross_entropy(sim, lbl) + F.cross_entropy(sim.T, lbl)) / 2

# Main
def main():
    clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16").to(device)
    clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")

    with open(captions_train_path) as f: captions_train = json.load(f)
    with open(captions_val_path) as f: captions_val = json.load(f)

    reextract_missing_or_short(train_frames_root, captions_train,
                               video_feats_train_path, text_feats_train_path,
                               clip_model, clip_processor)
    reextract_missing_or_short(val_frames_root, captions_val,
                               video_feats_val_path, text_feats_val_path,
                               clip_model, clip_processor)

    print("üì¶ Loading pre-extracted CLIP features...")
    video_train = torch.load(video_feats_train_path, map_location="cpu")
    text_train  = torch.load(text_feats_train_path, map_location="cpu")
    video_val   = torch.load(video_feats_val_path, map_location="cpu")
    text_val    = torch.load(text_feats_val_path, map_location="cpu")

    print(f"Train videos: {len(video_train)} | Val videos: {len(video_val)}")

    train_loader = DataLoader(VideoTextFeatureDataset(video_train, text_train),
                              batch_size=batch_size, shuffle=True, collate_fn=collate_fn)
    val_loader   = DataLoader(VideoTextFeatureDataset(video_val, text_val),
                              batch_size=batch_size, shuffle=False, collate_fn=collate_fn)

    model = TemporalGRU().to(device)
    loss_fn = CLIPLoss().to(device)
    optim = torch.optim.AdamW(list(model.parameters()) + list(loss_fn.parameters()),
                              lr=lr, weight_decay=1e-4)

    best_val, patience_count = float("inf"), 0
    train_loss, val_loss = [], []

    for ep in range(max_epochs):
        model.train()
        tr = 0
        for v, t in tqdm(train_loader, desc=f"Train {ep}", leave=False):
            v, t = v.to(device).float(), t.to(device).float()
            l = loss_fn(model(v), t)
            optim.zero_grad(); l.backward(); optim.step()
            tr += l.item()
        tr /= len(train_loader)
        train_loss.append(tr)

        model.eval()
        vl = 0
        with torch.no_grad():
            for v, t in tqdm(val_loader, desc=f"Val {ep}", leave=False):
                v, t = v.to(device).float(), t.to(device).float()
                vl += loss_fn(model(v), t).item()
        vl /= len(val_loader)
        val_loss.append(vl)

        print(f"[Epoch {ep}] Train {tr:.4f} | Val {vl:.4f}")

        if vl < best_val - min_delta:
            best_val = vl; patience_count = 0
            torch.save(model.state_dict(), save_model_path)
            print(f"‚úÖ Saved model: {save_model_path}")
        else:
            patience_count += 1
            print(f"No improvement ({patience_count}/{patience})")
            if patience_count >= patience:
                print("‚õî Early stopping triggered."); break

    plt.figure(figsize=(8,5))
    plt.plot(train_loss, label="Train Loss", marker="o")
    plt.plot(val_loss, label="Val Loss", marker="s")
    plt.legend(); plt.grid(True)
    plt.title("GRU Adapter (Re-extracted Short/Missing Videos)")
    plt.show()

    print(f"Best Val Loss: {best_val:.6f}")

if __name__ == "__main__":
    main()
```

```python
# -------------- Random 32frame + GRU (Ï∂îÎ°† Î∞è ÏÑ±Îä• ÌôïÏù∏) ------------------------------

import os, json, re, torch, torch.nn as nn, torch.nn.functional as F
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
import warnings

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

base_dir = "/content/drive/MyDrive/VIDEO_RETRIEVAL/random"
test_frames_root   = os.path.join(base_dir, "Test_1kA_random32")
captions_test_path = os.path.join(base_dir, "captions_test_1k.json")

video_feats_test_path = os.path.join(base_dir, "clip_video_embeds_test_random32_fixed.pt")
text_feats_test_path  = os.path.join(base_dir, "clip_text_embeds_test_random32.pt")
model_path            = os.path.join(base_dir, "gru_adapter_random32_fixed.pth")

# Î≥ëÎ†¨ Ïù¥ÎØ∏ÏßÄ Î°úÎìú
def load_images_parallel(frame_paths, max_workers=8):
    def _load_one(path):
        try:
            return Image.open(path).convert("RGB")
        except:
            return None
    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        imgs = list(ex.map(_load_one, frame_paths))
    return [i for i in imgs if i is not None]

# CLIP Î™®Îç∏ Î°úÎìú
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")
clip_model.eval()


# Í∏∞Ï°¥ ÌîºÏ≤ò Î°úÎìú (ÏûàÏùÑ Í≤ΩÏö∞)
if os.path.exists(video_feats_test_path):
    existing_video_embeds = torch.load(video_feats_test_path, map_location="cpu")
    print(f" Loaded existing video features: {len(existing_video_embeds)}")
else:
    existing_video_embeds = []

if os.path.exists(text_feats_test_path):
    existing_text_embeds = torch.load(text_feats_test_path, map_location="cpu")
    print(f" Loaded existing text features: {len(existing_text_embeds)}")
else:
    existing_text_embeds = []

# Í∏∞Ï°¥ ÎπÑÎîîÏò§ Ïù¥Î¶ÑÎì§ Î°úÎìú
existing_folders = sorted([d for d in os.listdir(test_frames_root)
                           if os.path.isdir(os.path.join(test_frames_root, d))])
print(f" Total video folders: {len(existing_folders)}")

# Ïù¥ÎØ∏ Ï≤òÎ¶¨Îêú ÎπÑÎîîÏò§ Ïàò Í∞êÏßÄ (ÏûÑÎ≤†Îî© Í∞úÏàòÏôÄ Ìè¥Îçî Í∞úÏàò ÎπÑÍµê)
processed_count = len(existing_video_embeds)
if processed_count < len(existing_folders):
    print(f" Missing videos detected: {len(existing_folders) - processed_count}")
else:
    print(" All videos already processed.")

# ÎàÑÎùΩÎêú ÎπÑÎîîÏò§Îßå Ï∂îÏ∂ú (32ÌîÑÎ†àÏûÑ Í∏∞Ï§Ä)
@torch.no_grad()
def extract_missing_features(frames_root, captions_path,
                             clip_model, clip_processor,
                             existing_count,
                             video_out_path, text_out_path):
    with open(captions_path, "r") as f:
        captions_dict = json.load(f)

    video_embeds, text_embeds = [], []
    video_folders = sorted([d for d in os.listdir(frames_root)
                            if os.path.isdir(os.path.join(frames_root, d))])

    missing_videos = video_folders[existing_count:]
    print(f" Extracting features for {len(missing_videos)} missing videos (32 frames each)...")

    for vid in tqdm(missing_videos, desc="Extracting missing video feats", ncols=100):
        vid_path = os.path.join(frames_root, vid)
        frame_files = sorted([os.path.join(vid_path, f)
                              for f in os.listdir(vid_path)
                              if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
        if len(frame_files) == 0:
            continue

        #  32ÌîÑÎ†àÏûÑÎßå ÏÇ¨Ïö©
        if len(frame_files) > 32:
            frame_files = sorted(np.random.choice(frame_files, 32, replace=False))
        imgs = load_images_parallel(frame_files)
        if len(imgs) == 0:
            continue

        inputs = clip_processor(images=imgs, return_tensors="pt").to(device)
        feats = clip_model.get_image_features(**inputs)
        feats = F.normalize(feats, dim=-1).cpu()
        video_embeds.append(feats)

        # ÌÖçÏä§Ìä∏ Îß§Ïπ≠
        key = re.sub(r"[^0-9]", "", vid)
        cap = captions_dict.get(key, None)
        if cap is None:
            cap = ""
        if isinstance(cap, list):
            cap = cap[0]
        elif not isinstance(cap, str):
            cap = str(cap)

        t_inputs = clip_processor(text=cap, return_tensors="pt", truncation=True).to(device)
        t_feats = clip_model.get_text_features(**t_inputs)
        t_feats = F.normalize(t_feats, dim=-1).squeeze(0).cpu()
        text_embeds.append(t_feats)

    # Í∏∞Ï°¥ + ÏÉàÎ°ú Ï∂îÍ∞ÄÎêú ÌîºÏ≤ò Ìï©ÏπòÍ∏∞
    if os.path.exists(video_out_path):
        old_videos = torch.load(video_out_path, map_location="cpu")
        video_embeds = old_videos + video_embeds
    if os.path.exists(text_out_path):
        old_texts = torch.load(text_out_path, map_location="cpu")
        text_embeds = old_texts + text_embeds

    torch.save(video_embeds, video_out_path)
    torch.save(text_embeds, text_out_path)
    print(f" Updated features saved: {len(video_embeds)} videos total")

# Ïã§Ìñâ
extract_missing_features(
    test_frames_root,
    captions_test_path,
    clip_model,
    clip_processor,
    len(existing_video_embeds),
    video_feats_test_path,
    text_feats_test_path
)

# ÌîºÏ≤ò Î°úÎìú ÎòêÎäî Ï∂îÏ∂ú
need_extract = not (os.path.exists(video_feats_test_path) and os.path.exists(text_feats_test_path))

if not need_extract:
    print(" Found cached features ‚Üí loading from disk.")
    video_embeds = torch.load(video_feats_test_path, map_location="cpu")
    text_embeds  = torch.load(text_feats_test_path, map_location="cpu")
    video_names  = sorted([d for d in os.listdir(test_frames_root)
                           if os.path.isdir(os.path.join(test_frames_root, d))])
else:
    print(" Cached features not found ‚Äî extracting...")
    video_embeds, video_names = extract_video_features(test_frames_root, clip_model, clip_processor)
    text_embeds = extract_text_features(captions_test_path, clip_model, clip_processor, video_names)

print(f" Loaded {len(video_embeds)} video feats and {len(text_embeds)} text feats")

# GRU Î™®Îç∏ Ï†ïÏùò
class TemporalGRU(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=1024, num_layers=1, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )
        self.proj = nn.Linear(hidden_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)

    def forward(self, x):
        if x.dim() == 2:
            x = x.unsqueeze(1)
        out, _ = self.gru(x)
        scores = self.attn(out)
        weights = F.softmax(scores, dim=1)
        pooled = (out * weights).sum(dim=1)
        out = self.proj(pooled)
        out = self.norm(out)
        return F.normalize(out, dim=-1)

# GRU Î™®Îç∏ Î°úÎìú
model = TemporalGRU().to(device)
state = torch.load(model_path, map_location=device)
model.load_state_dict(state)
model.eval()
print(f" Loaded trained GRU model from {model_path}")

# ÎπÑÎîîÏò§ Ïù∏ÏΩîÎî©
batch_size = 64
video_outputs = []

def pad_to_fixed_length(feat, target_len=32):
    """Î∂ÄÏ°±Ìïú ÌîÑÎ†àÏûÑÏùÄ 0ÏúºÎ°ú Ìå®Îî©, Ï¥àÍ≥ºÎ©¥ ÏïûÎ∂ÄÎ∂Ñ ÏûêÎ¶Ñ"""
    n = feat.size(0)
    if n == target_len:
        return feat
    elif n > target_len:
        return feat[:target_len]
    else:
        pad = torch.zeros(target_len - n, feat.size(1))
        return torch.cat([feat, pad], dim=0)

with torch.no_grad():
    for i in tqdm(range(0, len(video_embeds), batch_size), desc="GRU Encoding", ncols=100, leave=False):
        batch_feats = [pad_to_fixed_length(v) for v in video_embeds[i:i+batch_size]]
        v = torch.stack(batch_feats).to(device).float()
        v_out = model(v)
        video_outputs.append(v_out.cpu())

video_mat = torch.cat(video_outputs, dim=0)
text_mat  = F.normalize(torch.stack(text_embeds), dim=-1)


# Retrieval ÌèâÍ∞Ä (Video‚ÜíText, Text‚ÜíVideo)
video_mat = F.normalize(video_mat, dim=-1)
text_mat  = F.normalize(text_mat, dim=-1)
sim = video_mat @ text_mat.T

def recall_at_k(ranks, k):
    correct = torch.arange(ranks.size(0)).unsqueeze(1)
    hits = (ranks[:, :k] == correct).any(dim=1).float()
    return hits.mean().item()

def median_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:,1]
    return pos.float().median().item() + 1

def mean_rank(ranks):
    gt = torch.arange(ranks.size(0))
    pos = (ranks == gt.unsqueeze(1)).nonzero(as_tuple=False)[:,1]
    return pos.float().mean().item() + 1

# --- Video‚ÜíText ---
ranks_v2t = sim.argsort(dim=1, descending=True)
R1_v2t, R5_v2t, R10_v2t = recall_at_k(ranks_v2t,1), recall_at_k(ranks_v2t,5), recall_at_k(ranks_v2t,10)
MedR_v2t, MnR_v2t = median_rank(ranks_v2t), mean_rank(ranks_v2t)

# --- Text‚ÜíVideo ---
ranks_t2v = sim.argsort(dim=0, descending=True).T
R1_t2v, R5_t2v, R10_t2v = recall_at_k(ranks_t2v,1), recall_at_k(ranks_t2v,5), recall_at_k(ranks_t2v,10)
MedR_t2v, MnR_t2v = median_rank(ranks_t2v), mean_rank(ranks_t2v)

print("\n Retrieval Performance (Random 32f, GRU-based)")
print("‚îÄ‚îÄ Video ‚Üí Text ‚îÄ‚îÄ")
print(f"R@1 : {R1_v2t*100:.2f}%")
print(f"R@5 : {R5_v2t*100:.2f}%")
print(f"R@10: {R10_v2t*100:.2f}%")
print(f"MedR: {MedR_v2t:.1f}")
print(f"MnR : {MnR_v2t:.1f}")

print("\n‚îÄ‚îÄ Text ‚Üí Video ‚îÄ‚îÄ")
print(f"R@1 : {R1_t2v*100:.2f}%")
print(f"R@5 : {R5_t2v*100:.2f}%")
print(f"R@10: {R10_t2v*100:.2f}%")
print(f"MedR: {MedR_t2v:.1f}")
print(f"MnR : {MnR_t2v:.1f}")
```
